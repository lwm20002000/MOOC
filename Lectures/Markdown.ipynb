<h1>Max Drawdown</h1>
- Max loss you couldve experienced to be unlucky enough to buy the asset at its very peak and sold at the very bottom
- The worst possible loss within a timeframe.
- Worst case measure of downside risk

<h2>Constructing a wealth index</h2>
<p>Wealth Index: Take a thousand dollars and invest over time.

Compute the peaks - look at the prior peaks at any point in time.

The draw down is the distance between the blue line and the green line. - The unrealised gain (unsold but the feel of loss)
</p>
